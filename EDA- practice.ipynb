```python
country= pd.read_excel('Country-Code.xlsx')
country
```

```python
df= pd.merge(zomato,country,on='Country Code', how='left')
df
```

```python
df.Country.value_counts()
```

```python
country_values=df.Country.value_counts().values
country_values
```

```python
country_index=df.Country.value_counts().index
country_index
```

```python
plt.pie(country_values[:3],labels=country_index[:3],autopct='%1.2f%%')
```
